```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Nutrient_recommendation.csv')
df
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df_rice = df[df['Crop']=='rice']
df_rice
```

```python
crop_names = df['Crop']
nitrogen = df['Label_N']
phosphorus = df['Label_P']
potassium = df['Label_K']
temperature = df['Temperature']
humidity = df['Humidity']
rainfall = df['Rainfall']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(crop_names, nitrogen, marker='o', label='Nitrogen')
plt.plot(crop_names, phosphorus, marker='o', label='Phosphorus')
plt.plot(crop_names, potassium, marker='o', label='Potassium')

plt.xlabel('Crop')
plt.ylabel('Nutrient Requirement')
plt.title('Crop vs. NPK Requirements')
plt.xticks(rotation=90)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(crop_names, temperature, color='skyblue')
plt.xlabel('Crop')
plt.ylabel('Temperature (Â°C)')
plt.title('Crop vs. Temperature')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(crop_names, humidity, color='lightgreen')
plt.xlabel('Crop')
plt.ylabel('Humidity')
plt.title('Crop vs. Humidity')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(crop_names, rainfall, color='lightcoral')
plt.xlabel('Crop')
plt.ylabel('Rainfall (mm)')
plt.title('Crop vs. Rainfall')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
