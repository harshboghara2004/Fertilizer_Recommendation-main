```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Fertilizer Prediction.csv')
df
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
temperature = df['Temperature']
humidity = df['Humidity']
moisture = df['Moisture']
soil_type = df['Soil Type']
crop_type = df['Crop Type']
nitrogen = df['Nitrogen']
potassium = df['Potassium']
phosphorous = df['Phosphorous']
fertilizer_name = df['Fertilizer Name']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(temperature, fertilizer_name, color='skyblue')
plt.xlabel('Temperature')
plt.ylabel('Fertilizer Name')
plt.title('Temperature vs. Fertilizer Name')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(humidity, fertilizer_name, color='lightgreen')
plt.xlabel('Humidity')
plt.ylabel('Fertilizer Name')
plt.title('Humidity vs. Fertilizer Name')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(moisture, fertilizer_name, color='lightcoral')
plt.xlabel('Moisture')
plt.ylabel('Fertilizer Name')
plt.title('Moisture vs. Fertilizer Name')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
crop_types = df['Crop Type'].unique()
for crop_type in crop_types:
    plt.scatter(df[df['Crop Type']==crop_type]['Crop Type'], df[df['Crop Type']==crop_type]['Fertilizer Name'], label=crop_type)
plt.xlabel('Crop Type')
plt.ylabel('Fertilizer Name')
plt.title('Crop Type vs. Fertilizer Name')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(nitrogen, phosphorous, potassium, c='r', marker='o')

ax.set_xlabel('Nitrogen')
ax.set_ylabel('Phosphorus')
ax.set_zlabel('Potassium')
ax.set_title('NPK Content vs. Fertilizer Name')

for i, txt in enumerate(fertilizer_name):
    ax.text(nitrogen[i], phosphorous[i], potassium[i], txt, size=8, zorder=1)

plt.show()
```
